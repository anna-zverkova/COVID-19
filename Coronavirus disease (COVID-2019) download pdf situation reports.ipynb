```python
import numpy as np 
import pandas as pd 
import requests
from bs4 import BeautifulSoup
import shutil
import urllib.request
from urllib.request import urlretrieve
import wget
import time
import os
import fnmatch
```

```python
print('Beginning file download with requests')

bs = BeautifulSoup

url = 'https://www.who.int/emergencies/diseases/novel-coronavirus-2019/situation-reports/'
print("")

suffix = ".pdf"

link_list = []
```

```python
# Gets URL from user to scrape
response = requests.get(url, stream=True)
soup = bs(response.text)

for link in soup.find_all('a'): 
     if suffix in str(link): # If the link ends in .pdf
            link_list.append(link.get('href'))
print(link_list)            
```

```python
for link in link_list:
    full_link = 'https://www.who.int' + link
    print(full_link)
    urlretrieve(full_link)
```

```python
from requests import get


for link in link_list:
    full_link = 'https://www.who.int' + link

    # get request
    response = get(full_link)
    # write to file
    wget.download(full_link)
    time.sleep(5)
```

```python
# listing downloaded files

os.listdir()
```

```python
# we want to remove duplicates

#first we need to identify duplicates
# if file is duplicated system ads (1) or (2) or (3) at the end of the name


for file in os.listdir('.'):
    if fnmatch.fnmatch(file, '*).pdf'):
        print(file)
```

```python
# now lets delete duplicate pdf's

for file in os.listdir('.'):
    if fnmatch.fnmatch(file, '*).pdf'):
        print(file)
        os.remove(file)
```
